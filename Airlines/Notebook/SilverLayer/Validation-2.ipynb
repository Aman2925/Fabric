```python
def check_distinct_columns(table_name, columns):
    for col in columns:
        print("\n" + "=" * 100)
        print(f"{col}")
        print("=" * 100)

        query = f"""
        SELECT
            {col},
            COUNT(*) AS RecordCount
        FROM {table_name}
        GROUP BY {col}
        ORDER BY RecordCount DESC
        """

        display(spark.sql(query))
```

```python
from pyspark.sql.functions import col, count, when

def check_nulls(df, columns):
    
    expr = [
        count(
            when(
                col(c).isNull() | (col(c) == ""),
                c
            )
        ).alias(c)
        for c in columns
    ]

    df.select(*expr).show(truncate=False)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

window_spec = Window.partitionBy(
    "AirportID",
    "ObservationDate",
    "ObservationTime"
).orderBy(col("SeverityRank").desc())
```

```python
weather_valid = (
    weather_valid
    .withColumn("rn", row_number().over(window_spec))
    .filter(col("rn") == 1)
    .drop("rn")
)
```

```python
weather = weather_valid.unionByName(weather_null)
```

```python
weather = weather.drop("SeverityRank")
```
